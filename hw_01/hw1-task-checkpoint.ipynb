```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

#TODO
```

```python
# Функция по значению LIMIT_BAL
# возвращает буквенный класс
def class_limit(ball):
    if    ball < 0:
        return 'Ошибка данных'
    elif  ball   <= 10000:
        return 'A'
    elif ball <= 100000:
        return 'B'
    elif ball <= 200000:
        return 'C'
    elif ball <= 400000:
        return 'D' 
    elif ball <= 700000:
        return 'E'
    else:
        return 'F'
#
# Добавляем столбец LIMIT_CLASS_BALL:

for i in df['LIMIT_BAL']:
    df['LIMIT_CLASS_BALL']= class_limit(i)
    
df.head()
  
```

```python
df_edu_limit = df[['EDUCATION', 'LIMIT_BAL','SEX']]
df_edu_women = df_edu_limit.loc[df_edu_limit['SEX'] == 2]

# Возвращает среднее значение кредитного лимита
# женщин с образованием n (0 - 6)
def Limit_Ball_Average (n):
    df_edu_n_women = df_edu_women.loc[df_edu_women['EDUCATION'] == n]
    k = 1
    average = 0
    for i in df_edu_n_women['LIMIT_BAL']:
        delta = i
        k += 1
        average += delta
    return average/k
#
# Создаем дата фрейм 
df_plot_w_e_l = pd.DataFrame({
        'EDUCATION': [0,1,2,3,4,5,6],
        'LIMIT_AVERAGE': [Limit_Ball_Average (0),Limit_Ball_Average (1),Limit_Ball_Average (2),
                          Limit_Ball_Average (3),Limit_Ball_Average (4),Limit_Ball_Average (5),
                          Limit_Ball_Average (6)]
                           })
# Рисуем
df_plot_w_e_l.plot (x='EDUCATION', y= 'LIMIT_AVERAGE',   figsize=(15,12))
plt.title('Зависимость кредитного лимита и образования у женщин');
plt.xlabel('Образование');
plt.ylabel('Средний кредитный лимит');

```
